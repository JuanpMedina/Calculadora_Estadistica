```python
#Tkinter
from tkinter import *
import tkinter as tk
from tkinter import filedialog as fd

#Pandas
import pandas as pd

#Numpy
import numpy as np

#Matplotlib
from matplotlib import pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
```

```python
def main():
    global frameGrafico, ventana
    ventana = tk.Tk()
    ventana.geometry("600x600")
    ventana.title('Regresion Lineal')

    botonSubir = tk.Button(
        ventana, 
        text = "Subir archivo", 
        command = abrirArchivo,
        width = 600, height = 3
    )
    botonSubir.pack()

    frameGrafico = Frame(
        ventana, 
        bg = "lightblue", 
        width = 600, height = 494
    )
    frameGrafico.pack()
    
    ventana.mainloop()
```

```python
def abrirArchivo():
    archivo = fd.askopenfilename(
        title="abrir",
        initialdir="C:/",
        filetypes=(("Hoja de c√°lculo de Microsoft Excel","*.xlsx"),("Todos los archivos","*.*"))
    )
    
    datos = pd.read_excel(archivo)
    
    x = np.array(datos.iloc[:,0])
    y = np.array(datos.iloc[:,1])
    
    regresionLineal(x,y,datos)
    
def graficarDatos(x, y, yAjustado, datos):
    fig, ax = plt.subplots(
        dpi = 100,
        facecolor = 'white',
    )
    ax.grid()
    ax.scatter(x,y)
    ax.plot(x,yAjustado, 'r')
    plt.xlabel(datos.columns[0])
    plt.ylabel(datos.columns[1])
    
    canvas = FigureCanvasTkAgg(fig, master = frameGrafico)
    canvas.draw()
    canvas.get_tk_widget().grid(column=1, row=0, rowspan=5)
    
def regresionLineal(x,y,datos):
    n = len(x)
    sumaXiYi = sum([xi*yi for xi,yi in zip(x,y)])
    sumaXi = sum(x)
    sumaYi = sum(y)
    sumaXi2 = sum(x*x)
    sumaXi22 = sum(x)**2

    a1 = ((n*sumaXiYi)-(sumaXi*sumaYi)) / ((n*sumaXi2)-(sumaXi22))
    a0 = np.mean(y) - (a1*np.mean(x)) #np.mean para sacar la media.
    modelo = lambda x: a1*x + a0
    
    St = sum((y - np.mean(y))**2)
    Sr = sum([(a-b)**2 for a,b in zip(y,modelo(x))])
    R2 = (St - Sr) / St
    R = np.sqrt(R2)
    
    graficarDatos(x,y,modelo(x),datos)
    
    modeloStr = ('y=', a1, 'x +', a0)
    r2Str = ("El coeficiente R2= ", R2)
    rStr = ("El coeficiente R= ", R)
    
    labelModelo = tk.Label(
        ventana,
        text = modeloStr,
        font=("Verdana",15)
    )
    labelModelo.pack()
    
    labelR2 = tk.Label(
        ventana,
        text = r2Str,
        font=("Verdana",15)
    )
    labelR2.pack()
    
    labelR = tk.Label(
        ventana,
        text = rStr,
        font=("Verdana",15)
    )
    labelR.pack()
```

```python
main()
```
